```python
df["age"] = [22, 25, 29, 24, 30, 35, 40, 36, 28, 22, 23, 25, 26, 24, 31, 29, 27, 33, 35, 38]
df.head()
```

```python
plt.scatter(df["age"], df["Salary"],color = "orange")
plt.show()
```

```python
sort_age = df.sort_values("age")
```

```python
plt.plot(sort_age["age"], df["Salary"], color = "red", marker = "*", linestyle=":", linewidth="2")
plt.grid()
plt.show()
```

```python
plt.bar(sort_age["age"], df["Salary"], color="green")
plt.show()
```

```python
hr_sal = df[df["dept"] == "HR"]["Salary"]
hr_sal
```

```python
it_sal = df[df["dept"] == "IT"]["Salary"]
it_sal
```

```python
fin_sal = df[df["dept"] == "Finance"]["Salary"]
fin_sal
```

```python
plt.boxplot([hr_sal,it_sal,fin_sal], labels = ["HR","IT", "Finance"])
plt.show()
```

```python
salary_by_dept = df.groupby("dept")["Salary"].sum()
salary_by_dept
```

```python
plt.pie(salary_by_dept, labels = salary_by_dept.index, autopct = "%1.2f")
plt.axis("equal")
plt.show()
```

```python
hr_mean = sum(hr_sal/len(hr_sal))
it_mean = sum(it_sal/len(it_sal))
fin_mean = sum(fin_sal/len(fin_sal))
```

```python
plt.bar(["HR","IT","Finance"], [hr_mean, it_mean, fin_mean], color = ["green", "blue", "red"])
plt.show()
```

```python
df["experience"] = [1, 2, 4, 1.5, 5, 8, 12, 10, 4, 1, 1.2, 2, 3, 1.8, 5.5, 4.5, 3.2, 7, 8, 10]
df.head()
```

```python
plt.scatter(df["age"], df["Salary"], s = df['experience'], color = "orange", edgecolors = "black")
plt.show()
```

```python
plt.scatter(df["age"], df["Salary"], s = df['experience']*50, color = "orange", edgecolors = "black")
plt.title("Salary vs Age vs Experience")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python

plt.scatter(df['age'], df['Salary'], c = df["dept"].map({"HR": "yellow", "IT": "blue", "Finance": "red"}))
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Salary vs Age vs Dept")
plt.legend()
plt.show()
```

```python
color = {"HR": "yellow", "IT": "blue", "Finance": "red"}
for dept, color in color.items():
    df_dept = df[df["dept"] == dept]
    plt.scatter(df_dept["age"], df_dept["Salary"], c = color, label = dept)
plt.legend()
```

```python
flg, ax = plt.subplots(1,3, figsize = (10,5))
```

```python
ax[0].plot(sort_age["age"], df['Salary'], color = "red", marker  = "*", markersize = 2, linewidth = "2")
ax[0].grid()
ax[0].set_title("Line Plot")
ax[0].set_xlabel("Age")
ax[0].set_ylabel("Salary")
```

```python
flg, ax = plt.subplots(1,3, figsize = (10,5))
ax[0].plot(sort_age["age"], df['Salary'], color = "red", marker  = "*", markersize = 2, linewidth = "2")
ax[0].grid()
ax[0].set_title("Line Plot")
ax[0].set_xlabel("Age")
ax[0].set_ylabel("Salary")

ax[1].hist(df["Salary"], bins = 5, color = "skyblue")
ax[1].set_title("Histogram")
ax[1].set_xlabel("Salary")
ax[1].set_ylabel("Frequency")

ax[2].boxplot(df["Salary"])
ax[2].set_title("Bar Plot")
ax[2].set_xlabel("Salary")

plt.savefig("iamge.png")
plt.show()
```
