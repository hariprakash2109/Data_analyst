```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = [1,2,3]
y = [4,5,6]
```

```python
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
data = {
    "Salary" : [25000, 30000, 37000, 28000, 39000, 48000, 55000, 52000, 35000, 26000, 27000, 31000, 29000, 32000, 33000, 34000, 36000, 40000, 45000, 50000]
}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.shape
```

```python
plt.plot(df["Salary"])
plt.show()
```

```python
plt.plot(df["Salary"], color = "red", marker="o")
plt.grid()
plt.show()
```

```python
plt.plot(df["Salary"], color = "red", marker="o", linestyle = ":", linewidth = "2")
plt.grid()
plt.show()
```

```python
plt.hist(df["Salary"])
plt.show()
```

```python
plt.hist(df["Salary"],bins = 5, color = "green")
plt.show()
```
