```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
p = sns.load_dataset('penguins')
p.head()
```

```python
p['species'].value_counts()
```

```python
p['island'].value_counts()
```

```python
sns.scatterplot(data = p, x = "flipper_length_mm",y="body_mass_g",hue="island")
```

```python
sns.set_style("ticks")
sns.set_context("paper")
sns.scatterplot(data = p, x = "flipper_length_mm",y="body_mass_g",hue="island",palette="Set2")
sns.despine()
```

```python
sns.scatterplot(data=p,x="species",y="body_mass_g",hue="island",style="sex")
```

```python
columns = ["bill_length_mm","bill_depth_mm","flipper_length_mm","body_mass_g"]
p[columns].corr()
```

```python
sns.heatmap(data=p[columns].corr(),annot=True,cmap="Blues",vmin=-0.2,linecolor="black",linewidths=2)
```

```python
sns.rugplot(data=p,y="body_mass_g",hue="species",palette="Set2",height=0.2)
```

```python
sns.pairplot(data=p,hue="sex",palette="Set2",diag_kind="kde")
#sns.pairplot(data=p,hue="sex",palette="Set2",diag_kind="hist")
```

```python
graph=sns.PairGrid(data=p,hue="sex",palette="Set2")
graph.map_upper(sns.scatterplot)
graph.map_lower(sns.kdeplot)
graph.map_diag(sns.histplot)
graph.add_legend()
```
