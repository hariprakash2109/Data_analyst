```python
import sqlite3
con = sqlite3.connect(':memory:')
cur = con.cursor()
#drop = '''drop table studentdent'''
#Create Database & Tables — DDL
cur.execute('''CREATE TABLE student(id integer primary key autoincrement,name text not null,department text,age int)''')
```

```python
#Insert Records — DML
cur.executemany('''insert into student(name,department,age) values(?,?,?)''',
    [('Hari', 'MCA',22),
    ("Priya", "MBA", 24),
    ("John", "MCA", 23),
    ("Arun", "BCA", 21),
    ("Ram", "BCA", 20),
    ])
```

```python
#Fetch / Select Data — DQL
cur.execute( '''select * from student''')
for row in cur.fetchall():
  print(row)
```

```python
#WHERE Clause + Operators
cur.execute("select name,age from student where age>22")
print(cur.fetchall())
```

```python
#UPDATE Operations
cur.execute('update student set age=25 where name="Hari"')
```

```python
cur.execute("""
SELECT department, COUNT(*) AS total_studentdents
FROM student
GROUP BY department
""")
print(cur.fetchall())
```

```python
cur.execute("""
SELECT department, AVG(age)
FROM student
GROUP BY department
""")
print(cur.fetchall())
```

```python
cur.execute("""
SELECT department, MAX(age)
FROM student
GROUP BY department
""")
print(cur.fetchall())
```

```python
cur.execute("""
SELECT department, MIN(age)
FROM student
GROUP BY department
""")
print(cur.fetchall())
```

```python
cur.execute("""
SELECT department, GROUP_CONCAT(name, ', ')
FROM student
GROUP BY department
""")
print(cur.fetchall())
```

```python
cur.execute("""
CREATE TABLE marks (
    sid INTEGER,
    subject TEXT,
    mark INTEGER,
    FOREIGN KEY(sid) REFERENCES students(id)
)
""")

cur.executemany("""
INSERT INTO marks (sid, subject, mark)
VALUES (?, ?, ?)
""", [
    (1, "AI", 85),
    (1, "DBMS", 92),
    (2, "AI", 75),
    (3, "DBMS", 88),
])
```

```python
#JOINS (INNER JOIN)
cur.execute("""
SELECT s.id, s.name, m.subject, m.mark
FROM student s
INNER JOIN marks m ON s.id = m.sid
""")
print(cur.fetchall())
```

```python
cur.execute("""
SELECT s.id, s.name, m.subject, m.mark
FROM student s
LEFT JOIN marks m ON s.id = m.sid
""")
print(cur.fetchall())

```

```python
cur.execute("""
SELECT s.id, s.name, m.subject, m.mark
FROM marks m
LEFT JOIN student s ON s.id = m.sid
""")
print(cur.fetchall())
```

```python
cur.execute("""
SELECT s.id, s.name, m.subject, m.mark
FROM student s
LEFT JOIN marks m ON s.id = m.sid
UNION
SELECT s.id, s.name, m.subject, m.mark
FROM marks m
LEFT JOIN student s ON s.id = m.sid
""")
print(cur.fetchall())
```

```python
cur.execute("""
SELECT s.name, m.subject
FROM student s
CROSS JOIN marks m
""")
print(cur.fetchall())
```

**Join Type	    Supported in SQLite	      Explanation**
INNER JOIN	      ✅	                   Matching rows only
LEFT JOIN	        ✅	                   All (left table) + matches
RIGHT JOIN	      ❌	                   But can simulate
FULL OUTER JOIN	  ❌	                   Use UNION workaround
CROSS JOIN	      ✅	                   Cartesian product
